```python
np.random.seed(42)
data = pd.DataFrame({
    'Cement': np.random.normal(),
    'Blast_Furnace_Slag': np.random.normal(),
    'Fly_Ash': np.random.normal(),
    'Water': np.random.normal(),
    'Super_plasticizer': np.random.normal(),
    'Coarse_Aggregate': np.random.normal(1000, 50, 100),
    'Fine_Aggregate': np.random.normal(800, 40, 100),
    'Age': np.random.normal(45, 20, 100),
    'Concrete_compressive_strength': np.random.normal(35, 5, 100)
})

# Create subplots
fig, axes = plt.subplots(1, 9, figsize=(20, 5), constrained_layout=True)

# List of column names
columns = data.columns

# Loop through each column and create a boxplot in the corresponding subplot
for ax, col in zip(axes, columns):
    sns.boxplot(y=df[col], ax=ax)
    ax.set_title(col)
    ax.set_xlabel('')  # Remove x-label to avoid clutter
    ax.set_ylabel('')  # Remove y-label to avoid clutter

# Show the plot
plt.show()
```

```python
plt.figure(figsize = (25, 20))
plotnumber = 1

for col in df.columns:
    if plotnumber <= 9: 
        ax = plt.subplot(3, 3, plotnumber)
        sns.distplot(df[col])
        plt.xlabel(col, fontsize = 15)
        
    plotnumber += 1

plt.show()
```
